```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('heart.csv')
df.head()
```

```python
# a. Histogram
# Objective: Show the distribution of patient ages
plt.figure(figsize=(10, 8))
plt.hist(df['age'], bins=10)
plt.title('Histogram of Age\nObjective: Show distribution of patient ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# b. Dot Plot
# Objective: Visualize individual observations of maximum heart rate achieved
plt.figure()
y = np.zeros(len(df))
x = df['thalach']
plt.plot(x, y, 'o')
plt.title("Dot Plot of Maximum Heart Rate Achieved\nObjective: Show each patient's max heart rate")
plt.xlabel('Max Heart Rate Achieved')
plt.yticks([])   # hide y-axis ticks since they're all zero
plt.show()
```

```python
# c. Bar Plot
# Objective: Compare age and cholesterol levels
plt.bar(df['age'], df['chol'])
  
plt.title("Age Vs Cholestrol")
  
# Setting the X and Y labels
plt.xlabel('age')
plt.ylabel('chol')
  
# Adding the legends
plt.show()
```

```python
# d. Line Chart
```

```python
sns.lineplot(x='age', y='target', data=df)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='target', y='thalach', data=df, )
sns.histplot(df['thalach'], bins=20, kde=True)
sns.stripplot(x='target', y='thalach', data=df)
plt.title('Boxplot with Histogram and Scatter Plot for thalach')
plt.show()
```
