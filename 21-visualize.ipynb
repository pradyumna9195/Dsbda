```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('adult_dataset.csv')
df.head()
```

```python
# a. Histogram
# Objective: Show the distribution of ages
plt.figure(figsize=(10, 8))
plt.hist(df['age'], bins=10)
plt.title('Histogram of Age\nObjective: Show distribution of ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# b. Dot Plot
# Objective: Visualize individual observations of educational no
plt.figure()
y = np.zeros(len(df))
x = df['educational-num']
plt.plot(x, y, 'o')
plt.title("Dot Plot of Educational no")
plt.xlabel('Educational Number')
plt.yticks([])   # hide y-axis ticks since they're all zero
plt.show()
```

```python
# c. Bar Plot
# Objective: Compare age and hoursperweek
plt.figure()
plt.bar(df['age'], df['educational-num'])
  
plt.title("Age Vs educational-num")
  
# Setting the X and Y labels
plt.xlabel('age')
plt.ylabel('educational-num')
  
# Adding the legends
plt.show()
```

```python
# d. Line Chart
sns.lineplot(x='age', y='hours-per-week', data=df)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='age', y='hours-per-week', data=df, )
sns.histplot(df['age'], bins=20, kde=True)
sns.stripplot(x='age', y='hours-per-week', data=df)
plt.title('Boxplot with Histogram and Scatter Plot')
plt.show()
```
