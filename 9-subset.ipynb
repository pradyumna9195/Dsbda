### b. Merge two subsets
### c. Sort Data on Page total likes
### d. Transposing Data

```python
import pandas as pd
df=pd.read_csv("dataset_Facebook.csv", sep=";")
df.head()
```

# a. Create data subsets for type of post

```python
df.Type.unique()
```

```python
photo_df = df[df.Type=="Photo"]
status_df = df[df.Type=="Status"]
link_df = df[df.Type == 'Link']
video_df = df[df.Type == 'Video']
```

# b. Merge two subsets

pd.merge(left dataframe,right dataframe, on(Column name (or list) present in both DataFrames to join on),how(type of join),suffixes	(Tuple of strings to append to overlapping column names (default=('_x', '_y')))

### Merge two subsets (Photo & Status) on 'Page total likes'

```python
merged_photo_status=pd.merge(
    photo_df,
    status_df,
    on="Page total likes",
    how="inner")
```

```python
merged_photo_status.head()
```

## e. Melting Data to long format

pd.melt(df, id_vars(identifier variables that dont change(dont melt),remain as it is), value_vars(These columns will be unpivoted), var_name(The column name for the melted variable names) ,value_name(The column name for the values )

### Melt data to long format

```python
melted = pd.melt(
    df,
    id_vars=["Type"],
    value_vars=["Page total likes", "Lifetime Post Total Reach"],
    var_name='Metric',
    value_name='Value'
)
```

```python
print("Melted Data (long format)")
print(melted)
```

## f. Casting data to wide format

pd.pivot_table( index, column, values)

### Cast (pivot) the melted data back to wide format 
### Using pivot_table with mean aggregation to handle duplicates

```python
wide = melted.pivot_table(
    index='Type', # These stay as rows
    columns='Metric', # These become column headers
    values='Value', # The values to fill in the table
    aggfunc='mean'  # Aggregation function (in case of duplicates)
)
```

```python
print(wide)
```
