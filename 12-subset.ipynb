```python
# Subset by Country (e.g. United-States)
us_df = df[df['native-country'] == 'United-States']

# Subset by Sex (e.g. Female)
female_df = df[df['gender'] == 'Female']

# Subset by Race (e.g. Black)
black_df = df[df['race'] == 'Black']
```

```python
# Step 3: Merge Two Subsets
# t. Merge two subsets (e.g., Female and Black)
```

```python
# Merge by index (intersection of Female and Black)
merged_df = pd.merge(female_df, black_df, how='inner')

# If you want to merge on a key or concatenation:
# Example of concatenation
# merged_df = pd.concat([female_df, black_df], axis=0)
```

```python
# Step 6: Melting Data to Long Format
# w. Melting
```

```python
# Melt into long format
melted = pd.melt(df,
                 id_vars=['income'],
                 value_vars=['age', 'educational-num', 'capital-gain', 'hours-per-week'],
                 var_name='Attribute',
                 value_name='Value')

# Pivot to wide format
wide = melted.pivot_table(index='income', columns='Attribute', values='Value', aggfunc='mean').reset_index()

# Optional prints
print("Data Loaded and Cleaned")
print(wide.head())
```
