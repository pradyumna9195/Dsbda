```python
df.columns = ["SepalLength", "SepalWidth", "PetalLength", "PetalWidth", "Species"]
# Show first 5 rows
print(df.head())
```

```python
# (g) Create Data Subsets for Different Species
```

```python
# Subsets for each species
setosa = df[df['Species'] == 'Iris-setosa']
versicolor = df[df['Species'] == 'Iris-versicolor']
virginica = df[df['Species'] == 'Iris-virginica']

# Display one of them
setosa.head()
```

```python
# (h) Merge Two Subsets (Setosa and Versicolor)
```

```python
# Merge using concat
merged_df = pd.concat([setosa, versicolor], ignore_index=True)

# Show merged DataFrame
merged_df.head()
```

```python
# (i) Sort Data by Petal Length
```

```python
# Sort in ascending order
sorted_df = df.sort_values(by='PetalLength')

# Display sorted data
sorted_df.head()
```

```python
# (j) Transpose the Data
```

```python
# Transpose entire DataFrame
transposed_df = df.T  # T is shorthand for transpose

# Display transposed data
transposed_df.head()
```

```python
# (k) Melting Data to Long Format
```

```python
# Step 1: Melt the data to long format
melted_df = pd.melt(df, id_vars=['Species'], 
                    value_vars=['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth'], 
                    var_name='Feature', value_name='Value')

# Show melted data
melted_df.head()
```

```python
# (l) Casting Data to Wide Format (from Melted)
```

```python
# Step 2: Melt the DataFrame
melted_df = pd.melt(df, 
                    id_vars=['Species'], 
                    value_vars=['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth'], 
                    var_name='Feature', 
                    value_name='Value')

# Step 3: Pivot back to wide format
# Note: Use repeated group number (0, 1, 2...) for each row in original df
melted_df['RowID'] = melted_df.groupby('Feature').cumcount()

casted_df = melted_df.pivot(index='RowID', columns='Feature', values='Value')

# Step 4: Add 'Species' back
casted_df['Species'] = df['Species'].values

# Step 5: View the result
print(casted_df.head())
```
