```python
# a. Create Data Subsets by Area Affected
```

```python
# Categorize area into classes
def classify_area(area):
    if area == 0:
        return 'NotAffected'
    elif area <= 10:
        return 'PartiallyAffected'
    else:
        return 'MostlyAffected'

# Apply the function
df['AffectedClass'] = df['area'].apply(classify_area)

# Create subsets
not_affected = df[df['AffectedClass'] == 'NotAffected']
partially_affected = df[df['AffectedClass'] == 'PartiallyAffected']
mostly_affected = df[df['AffectedClass'] == 'MostlyAffected']

print("Subset Counts:")
print(df['AffectedClass'].value_counts())
```

```python
 # b. Merge Two Subsets
```

```python
# Merge NotAffected and PartiallyAffected
merged_df = pd.concat([not_affected, partially_affected])
print("Merged Shape:", merged_df.shape)
```

```python
# c. Sort Data Using Temp, Wind, and Area
```

```python
# Sort by temp, wind, and area
sorted_df = df.sort_values(by=['temp', 'wind', 'area'], ascending=[False, False, False])
print(sorted_df[['temp', 'wind', 'area']].head())
```

```python
# e. Melting Data to Long Format
```

```python
# Melt with id_vars as categorical columns
melted = pd.melt(df, id_vars=['X', 'Y', 'month', 'day'], var_name='Variable', value_name='Value')
print(melted.head())
```

```python
# Step (e.1): Clean the 'Value' column (NEW - added before pivoting)
```

```python
# Convert 'Value' to numeric, non-convertible values become NaN
melted['Value'] = pd.to_numeric(melted['Value'], errors='coerce')

# Drop rows with NaN in 'Value' (i.e., non-numeric data like 'NotAffected')
melted = melted.dropna(subset=['Value'])
```

```python
 # f. Casting Data to Wide Format (Pivot)  - Step (f): Pivot to Wide Format (now it will work)
```

```python
pivoted = melted.pivot_table(index=['month', 'day'], columns='Variable', values='Value', aggfunc='mean')
print(pivoted.head())
```
