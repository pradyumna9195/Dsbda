```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.gridspec as gridspec
```

```python
df = pd.read_csv('heart.csv')
df.head()
```

```python
# a. Histogram
# Objective: Show the distribution of patient ages
plt.figure(figsize=(10, 8))
plt.hist(df['age'], bins=10)
plt.title('Histogram of Age\nObjective: Show distribution of patient ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# b. Pie Chart
# Objective: Show proportion of patients with and without heart disease
counts = df['target'].value_counts().sort_index()  # 0 = No Disease, 1 = Disease
labels = ['No Disease', 'Disease']
plt.figure()
plt.pie(counts, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart of Heart Disease Status\nObjective: Show proportion of patients with/without disease')
plt.show()
```

```python
# c. Box Plot
# Objective: Display median, quartiles, and outliers of cholesterol
plt.figure()
plt.boxplot(df['chol'])
plt.title('Box Plot of Cholesterol\nObjective: Display median, quartiles, and outliers')
plt.ylabel('Cholesterol (mg/dL)')
plt.xticks([1], ['Cholesterol'])
plt.show()
```

```python
# d. Scatter Plot
# Objective: Examine relationship between age and maximum heart rate achieved
plt.figure()
plt.scatter(df['age'], df['thalach'])
plt.title('Scatter Plot: Age vs Max Heart Rate\nObjective: Examine relation between age and max heart rate')
plt.xlabel('Age')
plt.ylabel('Max Heart Rate Achieved')
plt.show()
```

```python
# e. Scatter Plot with Box Plots

plt.figure(figsize=(12, 6))
sns.scatterplot(x='age', y='chol', data=df, color='red')
sns.boxplot(x='target', y='age', data=df)
plt.title('Scatter Plot with Boxplot for Age')
plt.xlabel('Heart Disease Presence (target)')
plt.ylabel('Age')
plt.show()
```
